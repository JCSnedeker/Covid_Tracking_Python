Covid Tracking Project Data Analysis

```
import pandas as pd
import numpy as np
```

```
df=pd.read_csv("https://raw.githubusercontent.com/JCSnedeker/Covid_Tracking_Python/main/Covid_Tracking_Data_12_2_20.csv")
```

```
df.head()
```

```
date=df["date"]
deaths=df["deathIncrease"]
hospitalizations=df["hospitalizedIncrease"]
tests=df["totalTestResultsIncrease"]
cases=df["positiveIncrease"]
```

```
pos_rate=100*cases/tests
```

```
pos_rate.head()
```

```
import plotly.express as px
```

```
scatter1=px.scatter(df,x="date",y="deathIncrease", title="Daily Deaths")
```

```
scatter1
```

```
import matplotlib.pyplot as plt
```

```
fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4)
ax1.plot(date, deaths)
ax1.set_title("Daily Deaths")
ax1.invert_xaxis()
ax2.plot(date, tests)
ax2.set_title("Daily Tests")
ax2.invert_xaxis()
ax3.plot(date, hospitalizations)
ax3.set_title("Daily Hospitalizations")
ax3.invert_xaxis()
ax4.plot(date, cases)
ax4.set_title("Daily Cases")
ax4.invert_xaxis()

plt.rcParams["figure.figsize"] = [24,9]
plt.show()
```

```
fig, (ax5) = plt.subplots()
ax5.plot(date, pos_rate)
ax5.set_title("Positive Rate")
ax5.invert_xaxis()
plt.rcParams["figure.figsize"] = [6,9]
plt.show()
```
